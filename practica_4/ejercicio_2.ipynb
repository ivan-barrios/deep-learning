# b)
* 1. Error cometido en la predicción:
Durante el entrenamiento, la red neuronal compara la salida real (predicha) con la salida esperada (real), y el error es la diferencia entre ambas.
Incidencia en el cambio de los pesos: Cuanto mayor sea el error, mayor será el ajuste de los pesos, ya que la red intenta reducir ese error durante las iteraciones del entrenamiento. El signo de la modificación del peso depende de si el error es positivo o negativo. Esto se logra a través del algoritmo de descenso del gradiente, que ajusta los pesos para minimizar este error.
* 2. Valor de la derivada de la función de activación:
Las funciones de activación en las capas ocultas y de salida son importantes porque definen cómo se transforman las entradas en salidas dentro de las neuronas. En este caso, se usa la función "tanh" en la capa oculta y "sigmoid" en la capa de salida.
Incidencia en el cambio de los pesos: La derivada de la función de activación se utiliza en el cálculo del gradiente durante el retropropagado del error (backpropagation). Esto afecta cuánto cambia un peso en función de cómo responde la neurona a un cambio en su entrada. Si la derivada es muy pequeña, el cambio en los pesos será pequeño (este fenómeno se llama "vanishing gradient" en algunos casos). Si es grande, el cambio será mayor.
* 3. Valores anteriores de los pesos de la red:
En cada iteración del entrenamiento, los pesos se ajustan en función de los errores calculados en esa iteración, pero también tienen en cuenta los valores anteriores de los pesos.
Incidencia en el cambio de los pesos: Si el peso anterior ya está cerca de un valor óptimo, los cambios serán más pequeños. Si está lejos del valor óptimo, los cambios serán mayores. Este factor también incluye el concepto de "momentum" (si se utiliza), que evita que los pesos cambien drásticamente y ayuda a suavizar el proceso de ajuste.
