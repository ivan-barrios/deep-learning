```python
from sympy import symbols, diff
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy import lambdify
```

```python
import sympy as sp

#Definimos variables
j, k = sp.symbols('j k')

#Defino funcion
P = k**2 - 2*j*k + 2*j**2 + 18*j - 20*k + 201

display(P)

# calculo las derivadas parciales
dP_dj = sp.diff(P, j)
dP_dk = sp.diff(P, k)

display(dP_dj)
display(dP_dk)
```

```python
# Malla de puntos para graficar
J = np.linspace(-4, 4, 100)
K = np.linspace(-4, 4, 100)
J, K = np.meshgrid(J, K)
Z = lambdify((j, k), P, 'numpy')(J, K)

# Graficar
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.view_init(elev=20, azim=150)  # Puedes ajustar los valores de elev y azim para cambiar la vista
ax.plot_surface(J, K, Z, cmap='viridis', alpha=0.9)

ax.set_ylabel('K')
ax.set_zlabel('Z')
plt.show()
```

```python
#Para minimizar el precio de la vivienda necesitamos ver primero el vector gradiente, y como nos dan los
# signos de las respectivas variables para poder realizar el movimiento de x e y

#Inicializamos valores
alpha = 0.1
n_iterations = 300
pj=0
pk=0

for i in range (n_iterations):
    #Primero vemos en que sentido crece la funcion con los grandientes
    grad_j = dP_dj.evalf(subs={j:pj, k:pk})
    grad_k = dP_dk.evalf(subs={j:pj, k:pk})

    #actualizamos los valores de pj y pk para acercarnos a nuestro valor
    pj = pj - alpha * grad_j
    pk = pk - alpha * grad_k

    P_val = P.evalf(subs={j:pj, k:pk})

    print(f"Iterations nmbr {i + 1}, P(j,k) = {P_val}")

print(f"\n Con los valores j={pj} k={pk} obtenemos el minimo P(j,k) = {P_val}")



```
