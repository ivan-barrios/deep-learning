```python
#Calcularemos las derivadas parciales nosotros con python:
import sympy as sp

x, y = sp.symbols('x y')

#defino la funcion
f = x/(2*x**2 + 3*y**2 + 1)

# calculo las derivadas parciales
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

display(df_dx)
display(df_dy)
```

```python
# Evaluar las derivadas en el punto (x=1, y=-1)
valor_gradiente_x = df_dx.evalf(subs={x: 1, y: -1})
valor_gradiente_y = df_dy.evalf(subs={x: 1, y: -1})

vector_gradiente = (valor_gradiente_x, valor_gradiente_y)
print(f"\nEl vector gradiente en el punto (1, -1) es: {vector_gradiente}")

valor_gradiente_x = df_dx.evalf(subs={x: 0.2, y: -1})
valor_gradiente_y = df_dy.evalf(subs={x: 0.2, y: -1})

vector_gradiente = (valor_gradiente_x, valor_gradiente_y)
print(f"\nEl vector gradiente en el punto (0.2, -1) es: {vector_gradiente}")
```

# b)
* Para que la funcion crezca desde (0.2, -1), debemos incrementar los valores de x e y, ya que esto es lo que el gradiente en ese punto nos indica (Ambos positivo).

# c)

```python
px = 0.2
py = -1

# la funcin crece en direccin al gradiente
rx = px+0.1*df_dx.evalf(subs = {x:px,y:py})
ry = py+0.1*df_dy.evalf(subs = {x:px,y:py})

rz = f.evalf(subs = {x:rx, y:ry}) #se evalua la funcion en esos puntos

print("\nLa funcin f(x,y) = ", f)
print(" ")
print("En (%f,%f) vale %f" % (px, py, f.evalf(subs = {x:px,y:py})))
print("En (%f,%f) vale %f " % (rx, ry, rz) )
```

Vemos claramente como la funcion crece al aumentar x e y un 10%.

# d)
* Para que la funcion disminuya, debemos buscar la direccion opuesta (invirtiendo signos) del gradiente.

# e)

```python
px = 0.2
py = -1

# la funcin crece en direccin al gradiente
rx = px+0.1*(-1)*df_dx.evalf(subs = {x:px,y:py})
ry = py+0.1*(-1)*df_dy.evalf(subs = {x:px,y:py})

rz = f.evalf(subs = {x:rx, y:ry}) #se evalua la funcion en esos puntos

print("\nLa funcin f(x,y) = ", f)
print(" ")
print("En (%f,%f) vale %f" % (px, py, f.evalf(subs = {x:px,y:py})))
print("En (%f,%f) vale %f " % (rx, ry, rz) )
```

Vemos como claramente decrece la funcion (toma un valor menor)

```python
#Podemos seleccionar un punto aleatorio inicial
px = 0.2
py = -1
#decidimos parametros
alpha = 0.1
max_iterations = 300

for i in range (max_iterations):
    #Calculamos el gradiente (separado en sus componentes x e y) en ese punto
    grad_x = df_dx.evalf(subs={x:px, y:py})
    grad_y = df_dy.evalf(subs={x:px, y:py})

    #Actualizamos la posicion utilizando el opuesto del gradiente (Donde decrece mas rapido)
    px = px - alpha*grad_x
    py = py - alpha*grad_y

    f_val = f.evalf(subs={x:px, y:py})

    print(f"Iteracion nro {i+1}, f(x,y) = {f_val}")
    #Repetimos hasta que sea suficientemente chico o que se terminen las iteraciones

print(f"/n Por lo tanto, tenemos que en x={px}, y={py} tenemos el minimo f(x,y)={f_val}")
```

Para el maximo de la funcion seria lo mismo, solo que cambiando el signo del alpha, que indica los saltos que hace en las coordenadas x e y.
