```python
# Generating a bar chart to compare AI regulatory frameworks in different regions.
import matplotlib.pyplot as plt
import numpy as np

# Data for comparison of AI regulation frameworks
temas = ['Guía y regulación', 'Gobernanza', 'Apoyo financiero', 'Facilitadores de IA y otros incentivos']
cantidad_iniciativas = [370, 705, 352, 524] 

# Creating the bar chart
fig, ax = plt.subplots()

# Bar chart with regulation strength
bars = ax.bar(temas, cantidad_iniciativas, color=['blue', 'green', 'orange'])

# Adding labels and title to the chart
ax.set_xlabel('Políticas y estrategias nacionales de inteligencia artificial')
ax.set_ylabel('Cantidad de iniciativas sobre el tema')
ax.set_ylim(0, 750)

plt.xticks(fontsize=6)

# Adding the value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 1, int(yval), ha='center', va='bottom')

# Save the figure to a file (this can be referenced later in the report)

# Show the plot
plt.show()
```
