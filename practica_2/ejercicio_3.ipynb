a) Luego de la numerización se calculó el coeficiente de correlación lineal entre los atributos EDAD y RIESGO
y se obtuvo como resultado -0.71. ¿Cómo debe interpretarse este valor?

El coeficiente de correlacion lineal negativo indica que a medida que a medida que uno aumenta, el otro disminuye. 



```python
w_edad = -0.076
w_riesgo = -0.071
b = 0.071

BAJO = 1
MEDIO = 2
ALTO = 3

min_edad=53
max_edad=60

#normalizacion lineal uniforme
def normalizar_edad(edad):
    return ((edad - min_edad)/(max_edad - min_edad))

#normalizacion lineal uniforme
def normalizar_riesgo(riesgo):
    return ((riesgo - BAJO)/(ALTO - BAJO))

def calcular_respuesta(edad, riesgo):
    y = (w_edad * normalizar_edad(edad)) + (w_riesgo * normalizar_riesgo(riesgo)) + b
    return "SI" if y>0 else "NO"

print(
calcular_respuesta(55, ALTO),
calcular_respuesta(56, MEDIO),
calcular_respuesta(58, MEDIO),
calcular_respuesta(56, BAJO),
calcular_respuesta(59, BAJO),
calcular_respuesta(57, MEDIO),
calcular_respuesta(60, BAJO),
calcular_respuesta(53, ALTO),
calcular_respuesta(59, MEDIO),
calcular_respuesta(55, MEDIO),
)
```
